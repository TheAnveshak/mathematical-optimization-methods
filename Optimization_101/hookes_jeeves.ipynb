```python
from sage.all import *
```

```python
x, y, z, a, b = var('x y z a b')
```

```python
def hookes_jeeves(f, init=(0.0, 0.0), step=1e-1, tol=1e-6, max_iter=100):
    t = var('t')
    base = vector(init)
    dirs = [vector([1.0, 0.0]), vector([0.0, 1.0])]
    
    for _ in range(max_iter):
        prev = base
        for d in dirs:
            f0 = float(f(*base))
            forward = float(f(*(base + step*d)))
            backward = float(f(*(base - step*d)))
            
            if forward < backward and forward < f0:
                search = f(*tuple(base + t*d))
                direction =  1
            elif backward < forward and backward < f0:
                search = f(*tuple(base - t*d))
                direction = -1
            else:
                continue

            
            sols = solve(search.diff(t) == 0, t)
            if not sols:
                continue
            try:
                rval = sols[0].rhs()
            except:
                rval = sols[0]
            delta = float(N(rval)) * direction
            base = base + delta*d
        
        pattern = base - prev
        
        if pattern.norm() < tol:
            break
        f0 = float(f(*base))
        forward = float(f(*(base + step*pattern)))
        backward = float(f(*(base - step*pattern)))
        if forward < backward and forward < f0:
            search = f(*tuple(base + t*pattern))
            direction =  1
        elif backward < forward and backward < f0:
            search = f(*tuple(base - t*pattern))
            direction = -1
        else:
            continue
        sols = solve(search.diff(t) == 0, t)
        if not sols:
            continue
        try:
            rval = sols[0].rhs()
        except:
            rval = sols[0]
        delta = float(N(rval)) * direction
        base = base + delta*pattern
    
    return tuple(base)
```

```python
func = lambda a, b: 6*a**2 - 6*a*b + 2*b**2 - a - 2*b
t0 = (0.0, 0.0)
opt = hookes_jeeves(func, init=t0)
print(f"Computed minimum at: {opt}")
```
