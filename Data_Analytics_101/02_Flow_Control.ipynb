### **1.   If Else Statements**



If and Else are self explanatory, they read as 'If X then execute code A Else execute code B'

Elif is a combination of else and if and is utilised to pass extra conditional checks between If and Else

Decision making is initialised by If, followed by required Elifs and then bookended by Else

```python
''' Checks if marks are below passing,
    If not Checks if marks are above good marks,
    If not then marks are within passing range '''


# try changing the value of the 'marks' variable for different outputs

marks = 50
passing_marks = 30
good_marks = 75


if marks < passing_marks:
  print("Fail")
elif marks > good_marks:
  print("Passed with distinction")
else:
  print("Passed")
```

Indentation is usually also automatically handled by your IDE

A statement without proper indentation will be read incorrectly by your interpreter and throw up and error


```python
# statement without proper indentation

a = 33
b = 200
c = b>a

if c:
    print("b is greater than a") # you will get an error
```

Most processes require a group of instructions to be repeatedly executed, known as a **loop**

In Conditional Loops, repeated iteration of a chunk of code continues unless a certain boolean condtion is meant

There are two types of conditional loops



*   For Loops
*   While Loops



### **1. For Loops**

For loops iterate over any iterable object i.e. lists, tuples, strings, dictionaries etc.

They utilise 'temporary variables', where a temporary name is used as a representation of the individual element in the iterable object

eg. when iterating over a list of fruits, we write 'for fruit in fruits execute code A'

For loops, just like if statements can also be nested within each other

```python
# Example 1: Iterating through a list
fruits = ["apple", "banana", "cherry"]
for i in fruits:
  print(i)
```

```python
# Example 2: Iterating through a string
message = "Hello World"
for letter in message:
  print(letter)
```

```python
# Example 3: Using range() to iterate a specific number of times
for ijk in range(5):  # Prints numbers 0 to 4
  print(ijk)
```

```python
# Example 4: Using range() with start, stop, and step
for i in range(2, 10, 2):  # Prints even numbers from 2 to 8
  print(i)
```

```python
# Example 5: Iterating through a dictionary
my_dict = {"name": "Alice", "age": 30, "city": "New York"}
for key in my_dict:
  print(key, ":", my_dict[key])
```

```python
# Example 6: Iterating through a dictionary (key-value pairs)
for key, value in my_dict.items():
  print(key, ":", value)
```

```python
# Example 7: Nested for loops (printing a multiplication table)
for i in range(1, 6):  # Outer loop (rows)
    for j in range(1, 6):  # Inner loop (columns)
        print(i * j, end="\t")  # Print the product, separated by tabs
    print()  # Move to the next line after each row
```

### **2. While Loops**

```python
# Example 1: Basic while loop
count = 0
while count < 5:
  print(count)
  count += 1
```

```python
# Example 2: While loop with a condition
number = 10
while number > 0:
  print(number)
  number -= 2
```

```python
# Example 3: While loop with break statement
i = 0
while i < 10:
  print(i)
  if i == 5:
    break  # Exit the loop when i is 5
  i += 1
```

```python
# Example 4: While loop with continue statement
j = 0
while j < 10:
  j += 1
  if j % 2 == 0:
    continue  # Skip even numbers
  print(j)
```

# Exception

The <b>try</b> block lets you test a block of code for errors.

The <b>expect</b> block lets you handle the error.

The <b>else</b> block lets you execute code when there is no error.

The <b>finally</b> block lets you execute code, regardless of the result of the try- and except blocks.

```python
try:
    number = int(input("Enter a number: "))
    result = 10 / number
    print(f"Result: {result}")
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")
except ValueError:
    print("Error: Invalid input. Please enter a number.")
finally:
    print("Eat 5 star do noting")

```

# Functions
A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.

A function can return data as a result.

for eg. A quadratic function of the form a*x^2 + b*x + c has one parameter x

```python
# prompt: basic function defining examples

# Example 1: A simple function that prints a greeting
def greet():
  """This function prints a simple greeting."""
  print("Hello!")

# Calling the function
greet()
```

```python
# Example 2: A function with one parameter
def greet_person(name):
  """This function greets the person passed in as a parameter."""
  print(f"Hello, {name}!")

# Calling the function with an argument
greet_person("Alice")
greet_person("Bob")
```

```python
# Example 3: A function with multiple parameters
def add_numbers(num1, num2):
  """This function adds two numbers and prints the result."""
  sum_result = num1 + num2
  print(f"The sum of {num1} and {num2} is {sum_result}")

# Calling the function with multiple arguments
add_numbers(5, 3)
add_numbers(10, -2)
```

```python
# Example 7: A function with variable number of keyword arguments (**kwargs)
def print_info(**kwargs):
  """This function prints key-value pairs passed as keyword arguments."""
  for key, value in kwargs.items():
    print(f"{key}: {value}")

# Calling the function with keyword arguments
print_info(name="Charlie", age=25, city="London")
print_info(item="Book", price=19.99)
```

```python
import random

# ðŸ“š Book Recommender System
def Book_Recommendation(genre):
    """Recommends a book based on genre. Includes some classics and hidden gems."""
    library = {
        "Sci-Fi": ["Dune", "Neuromancer", "The Left Hand of Darkness"],
        "Fantasy": ["Mistborn", "The Hobbit", "American Gods"],
        "Mystery": ["The Girl with the Dragon Tattoo", "Gone Girl", "In the Woods"],
        "Philosophy": ["The Republic", "Meditations", "Sapiens"],
        "Informatics": ["The Pragmatic Programmer", "Clean Code", "Introduction to Algorithms"]
    }
    return random.choice(library.get(genre, ["Genre not found."]))

print("Recommended Book (Sci-Fi):", Book_Recommendation("Sci-Fi"))
print("Recommended Book (Philosophy):", Book_Recommendation("Philosophy"))
```

```python
# ðŸ“¦ Decision Matrix - Should You Read the Book?
def Should_Read(rating, pages, genre):
    """
    Determines whether to read a book based on:
    - Rating (out of 10)
    - Pages (short: <300, medium: 300-600, long: >600)
    - Genre (Sci-Fi, Mystery, Philosophy, Fantasy, Informatics)
    """
    if rating >= 8 and pages < 400:
        return "Definitely Read It!"
    elif rating >= 6 and genre == "Informatics":
        return "Read for Knowledge!"
    elif pages > 600:
        return "Maybe Later, It's a Tome!"
    else:
        return "Pass for Now."

print("Should You Read This Book?", Should_Read(8, 250, "Sci-Fi"))
print("Should You Read This Book?", Should_Read(5, 700, "Mystery"))
```

# Area of Triangle

```python
import numpy as np
def Area(a,b,c):
    if a+b>c and a+c>b and b+c>a:
        s = (a+b+c)/2
        A = np.sqrt(s*(s-a)*(s-b)*(s-c))
        return A;
    elif a+b == c or b+c==a or a+c==b:
        return 0;
    else:
        return None

Area(1,3,3)
```

# Bisection method

```python
def bisection(f,x1,x2, maxitr = 10, tol = 1e-6):
    i = 1
    error = 1
    while i< maxitr and error>= tol:
        x3 = (x1+x2)/2
        error = abs(x3 - f(x3))
        if np.sign(f(x3)) == np.sign(f(x2)):
            x1,x2 = x1,x3;
        if np.sign(f(x3))== np.sign(f(x1)):
            x1,x2 = x3,x2
        i += 1
    return x3
```

# Monte Carlo Approximation of pi

```python
import matplotlib.pyplot as plt
from random import random

# Number of points to generate
n = 10000  # You can increase this number for more accuracy but slower computation
```

```python
# Generate random points within the unit square (x, y in [0, 1])
points = [(random(), random()) for _ in range(n)]

# Identify points inside the unit circle (x^2 + y^2 <= 1)
circle_points = [p for p in points if p[0]**2 + p[1]**2 <= 1]
outside_points = [p for p in points if p[0]**2 + p[1]**2 > 1]
```

```python
# Calculate approximation of pi using the ratio of points inside the circle
pi_approximation = 4 * len(circle_points) / len(points)
print("Approximated Pi:", pi_approximation)

# Plot the points
circle_x, circle_y = zip(*circle_points)
outside_x, outside_y = zip(*outside_points)
```

# Cubic Solver

```python
import math

def solve_cubic(a, b, c, d):
    if abs(a) < 1e-8:
        a, b, c = b, c, d
        if abs(a) < 1e-8:
            a, b = b, c
            return [] if abs(a) < 1e-8 else [-b / a]
        D = b * b - 4 * a * c
        return [] if D < 0 else [(-b + math.sqrt(D)) / (2 * a), (-b - math.sqrt(D)) / (2 * a)] if D > 0 else [-b / (2 * a)]

    p = (3 * a * c - b * b) / (3 * a * a)
    q = (2 * b * b * b - 9 * a * b * c + 27 * a * a * d) / (27 * a * a * a)
    roots = []

    if abs(p) < 1e-8:
        roots = [math.cbrt(-q)]
    elif abs(q) < 1e-8:
        roots = [0, math.sqrt(-p), -math.sqrt(-p)] if p < 0 else [0]
    else:
        D = q * q / 4 + p * p * p / 27
        if abs(D) < 1e-8:
            roots = [-1.5 * q / p, 3 * q / p]
        elif D > 0:
            u = math.cbrt(-q / 2 - math.sqrt(D))
            roots = [u - p / (3 * u)]
        else:
            u = 2 * math.sqrt(-p / 3)
            t = math.acos(3 * q / p / u) / 3
            k = 2 * math.pi / 3
            roots = [u * math.cos(t), u * math.cos(t - k), u * math.cos(t - 2 * k)]

    return [r - b / (3 * a) for r in roots]
```

# Fun with Pictures!

```python
import cv2
import numpy as np

ascii_chars = "@%#*+=-:. "

image = cv2.imread('my_img.jpg')

if image is None:
    raise ValueError("Image not found")

image_resized = cv2.resize(image, (100, 100), interpolation=cv2.INTER_AREA)

cv2.imshow('Resized Image', image_resized)

gray = cv2.cvtColor(image_resized, cv2.COLOR_BGR2GRAY)

# Normalize grayscale to map to ASCII
# Scale pixel values (0-255) to index into ascii_chars
ascii_image = [
    "".join(ascii_chars[pixel * (len(ascii_chars) - 1) // 255] for pixel in row)
    for row in gray
]

# Print ASCII image line by line
for line in ascii_image:
    print(line)
```
