```python
sns.scatterplot(x="bill_length_mm", y="body_mass_g", data=ds, hue="species")
plt.title("Bill Length v/s penguin Mass", size=20, color="Cyan")
```

```python
sns.histplot(x ="body_mass_g", data=ds, kde = True, hue = "species")
plt.title("Mass", size=20, color="red")
```
