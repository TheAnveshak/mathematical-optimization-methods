```python
class Library:
    def __init__(self, name, location, established_year, librarian_name):
        self.name = name
        self.location = location
        self.established_year = established_year
        self.librarian_name = librarian_name
        self.catalog = {}
        self.members = {}
        self.genres = {
            'fiction': ['fantasy', 'horror', 'epic', 'adventure', 'self-help', 'biography'],
            'non-fiction': ['science', 'maths', 'physics', 'IT', 'chemical engineering', 'chemistry', 'aerospace', 'computer science','cosmology','biology','mechanical engineering','electrical engineering']
        }
        self.book_counter = {}  # Tracks the total copies of each title

    def initialize_catalog(self, book_dict):
        """ Initializes the catalog with a predefined dictionary of books """
        for isbn, book_info in book_dict.items():
            genre = book_info['Genre'].lower()
            title = book_info['Title']
            author = book_info['Author']
            year = book_info['Year']
            price = book_info['Price Rupees']

            # Generate a book ID for each copy
            book_id = f"{genre[:3].upper()}-{random.randint(1000, 9999)}"
            self.catalog[book_id] = {
                'isbn': isbn,
                'title': title,
                'author': author,
                'genre': genre,
                'year': year,
                'price': price,
                'available': True
            }

            # Update the book counter
            key = f"{title.lower()}_{author.lower()}"
            self.book_counter[key] = self.book_counter.get(key, 0) + 1

        print(f"Initialized catalog with {len(book_dict)} books.")

    def add_book(self, title, author, genre, year_published,price, copies=1):
        if genre not in [g for sublist in self.genres.values() for g in sublist]:
            print(f"Genre '{genre}' is not recognized.")
            return

        # Add each copy as a separate entry with a unique ID
        for _ in range(copies):
            book_id = f"{genre[:3].upper()}-{random.randint(1000, 9999)}"
            self.catalog[book_id] = {
                'title': title,
                'author': author,
                'genre': genre,
                'year': year_published,
                'price': 0,  # Placeholder for price
                'available': True
            }

        # Update the counter for that title
        key = f"{title.lower()}_{author.lower()}"
        self.book_counter[key] = self.book_counter.get(key, 0) + copies
        print(f"Added {copies} copy(ies) of '{title}' by {author}.")

    def issue_book(self, title, author):
        key = f"{title.lower()}_{author.lower()}"
        if self.book_counter.get(key, 0) == 0:
            print(f"No copies of '{title}' by {author} available.")
            return

        for book_id, book_info in self.catalog.items():
            if book_info['title'].lower() == title.lower() and book_info['author'].lower() == author.lower() and book_info['available']:
                self.catalog[book_id]['available'] = False
                print(f"Issued '{title}' (ID: {book_id}) to a member.")
                return

        print(f"All copies of '{title}' are currently issued.")

    def register_member(self, name, age, affiliation):
        roll_no = f"M-{random.randint(1000, 9999)}"
        self.members[roll_no] = {
            'name': name,
            'age': age,
            'affiliation': affiliation
        }
        print(f"Registered member {name} (Roll No: {roll_no}).")
        return roll_no

    def display_books(self):
        """Display the catalog as a DataFrame"""
        if not self.catalog:
            print("No books in the catalog.")
            return

        book_data = {
            "Book ID": [],
            "Title": [],
            "Author": [],
            "Year": [],
            "Price": [],
            "Status": []
        }

        for book_id, info in self.catalog.items():
            status = "Available" if info['available'] else "Issued"
            book_data["Book ID"].append(book_id)
            book_data["Title"].append(info['title'])
            book_data["Author"].append(info['author'])
            book_data["Year"].append(info['year'])
            book_data["Price"].append(info['price'])
            book_data["Status"].append(status)

        df_books = pd.DataFrame(book_data)
        return df_books


    def display_members(self):
        """Display the members as a DataFrame"""
        if not self.members:
            print("No members registered.")
            return

        member_data = {
            "Roll No": [],
            "Name": [],
            "Age": [],
            "Affiliation": []
        }

        for roll_no, info in self.members.items():
            member_data["Roll No"].append(roll_no)
            member_data["Name"].append(info['name'])
            member_data["Age"].append(info['age'])
            member_data["Affiliation"].append(info['affiliation'])

        df_members = pd.DataFrame(member_data)
        return df_members


```

```python
def clean_price(p):
    if isinstance(p, str):
        p = p.lower().replace('rs', '').replace('.', '').replace(' ', '')
        if p.isdigit():
            return float(p)
        try:
            return float(p)
        except:
            return None
    elif isinstance(p, (int, float)):
        return float(p)
    else:
        return None

df['Price Rupees'] = df['Price Rupees'].apply(clean_price)

# Drop rows with invalid Price
df = df.dropna(subset=['Price Rupees'])

# Lowercase genres for easy checking
df['Genre'] = df['Genre'].str.lower()

# List of valid genres you expect
valid_genres = ['it', 'physics', 'biology', 'horror', 'maths', 'chemical engineering',
                'epic', 'cosmology', 'computer science', 'science', 'biography',
                'mechanical engineering']

# Keep only rows with genres in valid list, drop others
df = df[df['Genre'].isin(valid_genres)]

# Reset index after cleaning
df = df.reset_index(drop=True)

# Check cleaned dataframe
print(df.head())
print(df.info())
print("Null values left:\n", df.isnull().sum())
```

```python
# Remove rows with missing 'title' or 'authors'
df = df.dropna(subset=['title', 'authors'])

# Remove rows where average_rating is missing or not between 0 and 5
df = df[df['average_rating'].between(0, 5)]

# Remove rows with num_pages <= 0 or null
df = df[df['  num_pages'].notnull() & (df['  num_pages'] > 0)]

# Convert publication_date to datetime (ignore errors for now)
df['publication_date'] = pd.to_datetime(df['publication_date'], errors='coerce')

# Remove rows where publication_date could not be converted (NaT)
df = df.dropna(subset=['publication_date'])

# Clean 'language_code' column: keep only standard language codes like 'eng', 'spa', etc.
# Remove rows where language_code is null or unusual (optional)
df = df[df['language_code'].notnull()]

# Reset index for clean dataframe
df = df.reset_index(drop=True)

print("After cleaning:")
print(df.info())
```
